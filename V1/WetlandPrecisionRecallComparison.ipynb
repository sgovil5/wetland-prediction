```python
labels = ['Decision Tree with Gini', 'Decision Tree with Entropy', 'Bagging', 'Random Forest', 'Ada Boosting with SAMME.R', 'Ada Boosting with SAMME', 'Gradient Boosting', 'Bagging PCA', 'Extreme Gradient Boosting', 'Light Gradient Boosting', 'LDA', 'QDA', 'KNN', 'Logistic Regression'  ]
nonwetlands = [.8782, .8889, .9169, .9294, .9050, .9031, .9289, .8953, .9381, .9394, .9387, .9132, .8844, .9008 ]
wetlands = [.9064, .9068, .9184, .9116, 0.8629, .8994, .9247, .8788, .9246, .9244, .7640, .7951, .8205, .8095 ]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, nonwetlands, width, label='Non-wetland')
rects2 = ax.bar(x + width/2, wetlands, width, label='Wetland')

ax.set_ylabel('Precision')
ax.set_title('Comparing Models Precision Values for Wetland vs Non-Wetland')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(bbox_to_anchor=(.105,.24), loc="upper left",  bbox_transform=fig.transFigure)

fig.tight_layout()

ax.set_ylim([.40, 1])
plt.setp(ax.xaxis.get_majorticklabels(), rotation=270)
plt.grid(True, alpha=.5)

plt.show()
```
