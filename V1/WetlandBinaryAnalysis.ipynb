```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
import pandas as pd
import seaborn as sns
import SeabornFig2Grid as sfg

%matplotlib inline
sns.set_style('whitegrid')
sns.set(rc={"figure.dpi":70, 'savefig.dpi':70}) #defining dpi setting
sns.set_context('notebook')
sns.set_style("ticks")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
dfes = pd.read_csv('RiverRoad_Quan_RF.csv')
dfes
```

```python
def stacked_histograms(dfes, statistic, hue, title):
    fig, ax = plt.subplots(2, 3, figsize=(20, 12))
    sns.set_style("white")

    sns.histplot(data=dfes, ax=ax[0,0], stat="count", multiple="stack", x=f"{statistic}_MEAN", hue=hue, element="bars", legend=True).set_title(f"{title} Mean vs Wetland")

    sns.histplot(data=dfes,ax=ax[0,1], stat="count", multiple="stack", x=f"{statistic}_STD", hue=hue, element="bars", legend=True).set_title(f"{title} Standard Deviation vs Wetland")

    sns.histplot(data=dfes,ax=ax[0,2], stat="count", multiple="stack", x=f"{statistic}_RANGE", hue=hue, element="bars", legend=True).set_title(f"{title} Range vs Wetland")

    sns.histplot(data=dfes, ax=ax[1,0], stat="count", multiple="stack", x=f"{statistic}_MIN", hue=hue, element="bars", legend=True).set_title(f"{title} Minimum vs Wetland")

    sns.histplot(data=dfes, ax=ax[1,1], stat="count", multiple="stack", x=f"{statistic}_MAX", hue=hue, element="bars", legend=True).set_title(f"{title} Maximum vs Wetland")

    sns.histplot(data=dfes, ax=ax[1,2], stat="count", multiple="stack", x=f"{statistic}_SUM", hue=hue, element="bars", legend=True).set_title(f"{title} Sum vs Wetland")
```

```python
for (columnName, columnData) in dfes.iteritems():
    if("qu" in columnName and "MIN" in columnName):
        columnName = columnName.split('_')[0] + '_' + columnName.split('_')[1]
        stacked_histograms(dfes, columnName, 'wetland', columnName)
```
